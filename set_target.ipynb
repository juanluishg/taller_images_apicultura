```python
import pandas as pd
df = pd.read_csv("datasets/image_1.csv", index_col=0)
```

```python
df["clase"] = 0
```

```python
x = [461,582,564,576,548,982,1392,1328,1618,
     1742,1381,1232,1285,1544,2071,
     2179,2096,2155,2840,2538,2854,2666,3568,
     4122,3904,4043,4063,3816,4110,4910,5139,5250]
```

```python
y = [805,1136,1157,1351,1411,1865,184,357,495,
     671,858,1005,1031,1261,819,961,1285,
     1684,398,908,1545,2708,1085,675,1104,1150,1548,
     1831,2223,620,1272,1352]
```

```python
len(x)
```

```python
x_range = []
for i in x:
    x_range.append([i-25, i+25])
```

```python
y_range = []
for i in y:
    y_range.append([i-25, i+25])
```

```python
varroa = []
for i, p in df.iterrows():
    for ix, iy in zip(x_range, y_range): 
        if p.x >= ix[0] and p.x <= ix[1] and p.y >= iy[0] and p.y <= iy[1]:
            df.loc[i, "clase"] = 1
            print(i)
            varroa.append(i)
```
